# Import packages for plotting graphs and manipulating data

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plot the noisy data (x,y1) and clean data (x,y2)

```python
num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
    # f = np.sin(x) * (1 / (1 + np.exp(-x))) 
    f = np.abs(x) * np.sin(x)

    return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y2      = fun(x)
y1      = y2 + nn * std

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

# $\hat{\Theta}$ is solution of least square problem if $\|A\hat{\Theta}-y1\|^2\leq\|A\Theta-y1\|^2$ for any n-vector $\Theta$.
# We can find $\hat{\Theta}$ by this equation $\hat{\Theta}=(A^TA)^{-1}A^Ty1$

```python
A_0=np.zeros((num,1))
for i in range(num):
    A_0[i,0]=1
    
A_1=np.zeros((num,2))
for i in range(num):
    A_1[i,0]=1
    A_1[i,1]=x[i]
    
A_2=np.zeros((num,3))
for i in range(num):
    A_2[i,0]=1
    A_2[i,1]=x[i]
    A_2[i,2]=x[i]**2
    
A_3=np.zeros((num,4))
for i in range(num):
    A_3[i,0]=1
    A_3[i,1]=x[i]
    A_3[i,2]=x[i]**2
    A_3[i,3]=x[i]**3
    
A_4=np.zeros((num,5))
for i in range(num):
    A_4[i,0]=1
    A_4[i,1]=x[i]
    A_4[i,2]=x[i]**2
    A_4[i,3]=x[i]**3
    A_4[i,4]=x[i]**4
    
A_5=np.zeros((num,6))
for i in range(num):
    A_5[i,0]=1
    A_5[i,1]=x[i]
    A_5[i,2]=x[i]**2
    A_5[i,3]=x[i]**3
    A_5[i,4]=x[i]**4
    A_5[i,5]=x[i]**5
    
A_6=np.zeros((num,7))
for i in range(num):
    A_6[i,0]=1
    A_6[i,1]=x[i]
    A_6[i,2]=x[i]**2
    A_6[i,3]=x[i]**3
    A_6[i,4]=x[i]**4
    A_6[i,5]=x[i]**5
    A_6[i,6]=x[i]**6
    
A_7=np.zeros((num,8))
for i in range(num):
    A_7[i,0]=1
    A_7[i,1]=x[i]
    A_7[i,2]=x[i]**2
    A_7[i,3]=x[i]**3
    A_7[i,4]=x[i]**4
    A_7[i,5]=x[i]**5
    A_7[i,6]=x[i]**6
    A_7[i,7]=x[i]**7
    
A_8=np.zeros((num,9))
for i in range(num):
    A_8[i,0]=1
    A_8[i,1]=x[i]
    A_8[i,2]=x[i]**2
    A_8[i,3]=x[i]**3
    A_8[i,4]=x[i]**4
    A_8[i,5]=x[i]**5
    A_8[i,6]=x[i]**6
    A_8[i,7]=x[i]**7
    A_8[i,8]=x[i]**8
    
A_9=np.zeros((num,10))
for i in range(num):
    A_9[i,0]=1
    A_9[i,1]=x[i]
    A_9[i,2]=x[i]**2
    A_9[i,3]=x[i]**3
    A_9[i,4]=x[i]**4
    A_9[i,5]=x[i]**5
    A_9[i,6]=x[i]**6
    A_9[i,7]=x[i]**7
    A_9[i,8]=x[i]**8
    A_9[i,9]=x[i]**9
    
theta_0=((np.linalg.inv(A_0.T.dot(A_0))).dot(A_0.T)).dot(y1)
theta_1=((np.linalg.inv(A_1.T.dot(A_1))).dot(A_1.T)).dot(y1)
theta_2=((np.linalg.inv(A_2.T.dot(A_2))).dot(A_2.T)).dot(y1)
theta_3=((np.linalg.inv(A_3.T.dot(A_3))).dot(A_3.T)).dot(y1)
theta_4=((np.linalg.inv(A_4.T.dot(A_4))).dot(A_4.T)).dot(y1)
theta_5=((np.linalg.inv(A_5.T.dot(A_5))).dot(A_5.T)).dot(y1)
theta_6=((np.linalg.inv(A_6.T.dot(A_6))).dot(A_6.T)).dot(y1)
theta_7=((np.linalg.inv(A_7.T.dot(A_7))).dot(A_7.T)).dot(y1)
theta_8=((np.linalg.inv(A_8.T.dot(A_8))).dot(A_8.T)).dot(y1)
theta_9=((np.linalg.inv(A_9.T.dot(A_9))).dot(A_9.T)).dot(y1)
```

# Plot the polynomial curves that fit the noisy data by the least square error with varying p = 0~9, $\hat{f}(x)=\Theta_0x^0+\Theta_1x^1+\Theta_2x^2+...+\Theta_px^p$


```python
plt.plot(x, y1, 'b.', x, theta_0*np.ones((num,1)), 'r')
plt.title('p=0')
plt.show()
plt.plot(x, y1, 'b.', x, theta_1[0]+theta_1[1]*x, 'r')
plt.title('p=1')
plt.show()
plt.plot(x, y1, 'b.', x, theta_2[0]+theta_2[1]*x+theta_2[2]*x**2, 'r')
plt.title('p=2')
plt.show()
plt.plot(x, y1, 'b.', x, theta_3[0]+theta_3[1]*x+theta_3[2]*x**2+theta_3[3]*x**3, 'r')
plt.title('p=3')
plt.show()
plt.plot(x, y1, 'b.', x, theta_4[0]+theta_4[1]*x+theta_4[2]*x**2+theta_4[3]*x**3+theta_4[4]*x**4, 'r')
plt.title('p=4')
plt.show()
plt.plot(x, y1, 'b.', x, theta_5[0]+theta_5[1]*x+theta_5[2]*x**2+theta_5[3]*x**3+theta_5[4]*x**4+theta_5[5]*x**5, 'r')
plt.title('p=5')
plt.show()
plt.plot(x, y1, 'b.', x, theta_6[0]+theta_6[1]*x+theta_6[2]*x**2+theta_6[3]*x**3+theta_6[4]*x**4+theta_6[5]*x**5+theta_6[6]*x**6, 'r')
plt.title('p=6')
plt.show()
plt.plot(x, y1, 'b.', x, theta_7[0]+theta_7[1]*x+theta_7[2]*x**2+theta_7[3]*x**3+theta_7[4]*x**4+theta_7[5]*x**5+theta_7[6]*x**6+theta_7[7]*x**7, 'r')
plt.title('p=7')
plt.show()
plt.plot(x, y1, 'b.', x, theta_8[0]+theta_8[1]*x+theta_8[2]*x**2+theta_8[3]*x**3+theta_8[4]*x**4+theta_8[5]*x**5+theta_8[6]*x**6+theta_8[7]*x**7+theta_8[8]*x**8, 'r')
plt.title('p=8')
plt.show()
plt.plot(x, y1, 'b.', x, theta_9[0]+theta_9[1]*x+theta_9[2]*x**2+theta_9[3]*x**3+theta_9[4]*x**4+theta_9[5]*x**5+theta_9[6]*x**6+theta_9[7]*x**7+theta_9[8]*x**8+theta_9[9]*x**9, 'r')
plt.title('p=9')
plt.show()

```

# Check the error with this way $\sum_{j=1}^n r_j^2$ where $r_j=y1_j-\hat{f}(x_j)$ and plot the error 

```python
p=np.arange(0,10)
error=np.zeros(10)
error[0]=sum([r**2 for r in y1-(theta_0)])
error[1]=sum([r**2 for r in y1-(theta_1[0]+theta_1[1]*x)])
error[2]=sum([r**2 for r in y1-(theta_2[0]+theta_2[1]*x+theta_2[2]*x**2)])
error[3]=sum([r**2 for r in y1-(theta_3[0]+theta_3[1]*x+theta_3[2]*x**2+theta_3[3]*x**3)])
error[4]=sum([r**2 for r in y1-(theta_4[0]+theta_4[1]*x+theta_4[2]*x**2+theta_4[3]*x**3+theta_4[4]*x**4)])
error[5]=sum([r**2 for r in y1-(theta_5[0]+theta_5[1]*x+theta_5[2]*x**2+theta_5[3]*x**3+theta_5[4]*x**4+theta_5[5]*x**5)])
error[6]=sum([r**2 for r in y1-(theta_6[0]+theta_6[1]*x+theta_6[2]*x**2+theta_6[3]*x**3+theta_6[4]*x**4+theta_6[5]*x**5+theta_6[6]*x**6)])
error[7]=sum([r**2 for r in y1-(theta_7[0]+theta_7[1]*x+theta_7[2]*x**2+theta_7[3]*x**3+theta_7[4]*x**4+theta_7[5]*x**5+theta_7[6]*x**6+theta_7[7]*x**7)])
error[8]=sum([r**2 for r in y1-(theta_8[0]+theta_8[1]*x+theta_8[2]*x**2+theta_8[3]*x**3+theta_8[4]*x**4+theta_8[5]*x**5+theta_8[6]*x**6+theta_8[7]*x**7+theta_8[8]*x**8)])
error[9]=sum([r**2 for r in y1-(theta_9[0]+theta_9[1]*x+theta_9[2]*x**2+theta_9[3]*x**3+theta_9[4]*x**4+theta_9[5]*x**5+theta_9[6]*x**6+theta_9[7]*x**7+theta_9[8]*x**8+theta_9[9]*x**9)])

plt.plot(p,error,'o')
plt.xlabel('p')
plt.ylabel('error')
plt.title('Error when p=0~9')
plt.show()
```
